```python
print(aba['Sex'].value_counts())
sns.countplot('Sex', data=aba, palette='Set3')
plt.title("Sex Distribution")

plt.show()
```

```python
# Create a fig, ax
fig, ax = plt.subplots()

# Create a scatter plot with first 1000 data points
ax.scatter(aba["Diameter"][:1000], aba["Rings"][:1000], color="darkmagenta")

# Create a legend
ax.set(title="Diameter and Nr. of Rings",
       xlabel="Diameter",
       ylabel="Nr. of Rings");
```

```python
plt.figure(figsize=(10,5))
sns.distplot(aba['Age'])
plt.show()
```

```python
fig, axs = plt.subplots(3,5, figsize=(20,6),)
fig.subplots_adjust(hspace = 0.5, wspace=.2)
axs = axs.ravel()

for index, column in enumerate(aba.columns):
    axs[index-1].set_title("{} vs. Age".format(column),fontsize=16)
    axs[index-1].scatter(x=aba[column],y=aba['Age'],color='green',edgecolor='k')

fig.tight_layout(pad=1)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(aba.corr(),cmap='coolwarm',annot=True)
```
